```python
import pandas as pd
from scipy.stats import *
```

```python
data = {
    "18 or Less": [37, 213],
    "19 to 35": [48, 202],
    "36 to 54": [56, 194],
    "55 or Older": [73, 177]
}

index = ["Yes", "No"]
df = pd.DataFrame(data, index=index)
df
```

```python
proportion_df = df / df.sum()
proportion_df
```

```python
import math

p1 = 0.852
p2 = 0.708
p_diff = p1 - p2

n = 250
z_score = 1.96

se = math.sqrt((p1 * (1 - p1) / n) + (p2 * (1 - p2) / n))

lower = p_diff - z_score * se
upper = p_diff + z_score * se
print(f'Confidence interval [{lower:.3f};{upper:.3f}]')
```

The proportions of viewers differ across age groups and show a decreasing trend with increasing age. This suggests that viewing the 11 P.M. news may be related to age group
