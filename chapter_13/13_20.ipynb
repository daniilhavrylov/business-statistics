```python
import pandas as pd
from scipy.stats import *
```

```python
data = {
    "Below Industry Average": [4, 10],
    "Equal to Industry Average": [12, 4],
    "Above Industry Average": [8, 2],
}
index = ["No", "Yes"]

df = pd.DataFrame(data, index=index)
df
```

```python
chi2, p_value, dof, expected = chi2_contingency(df)

print("Chi-square:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

We do not have enough evidence to reliably use the chi-square test because one of the expected values is 4, which is below the recommended minimum of 5.


**Ho:** There is no difference in delivery-time performance between firms using computer-assisted ordering and those not using computer-assisted ordering.

**Ha:** Delivery-time performance differs between firms that use computer-assisted ordering and firms that do not.

**The chi-square test indicates that we have sufficient evidence to reject the null hypothesis at alpha = 0.05.**

```python
import matplotlib.pyplot as plt
row_pct = df.div(df.sum(axis=1), axis=0) * 100

row_pct.plot(kind="bar")
plt.ylabel("Percentage")
plt.title("Delivery-Time Performance by Ordering Method")
plt.show()
```
