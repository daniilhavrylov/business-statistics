```python
import pandas as pd
from scipy.stats import *
```

```python
data = {
    "Watch Tennis": [16, 4],
    "Do Not Watch Tennis": [24, 56]
}
index = ["Drink Wine", "Do Not Drink Wine"]

df = pd.DataFrame(data, index=index)
df
```

```python
row_percentages = df.divide(df.sum(axis=1), axis=0) * 100
column_percentages = df.divide(df.sum(axis=0), axis=1) * 100
cell_percentages = df / df.values.sum() * 100
```

```python
print("Row percentages (%)")
print(row_percentages.round(2))

print("\nColumn percentages (%)")
print(column_percentages.round(2))

print("Cell percentages (%)")
print(cell_percentages.round(2))
```

**Ho:** There is no relationship between wine consumption and whether a person watches professional tennis on television.

**Ha:** There is a relationship between wine consumption and whether a person watches professional tennis on television.

```python
chi2, p_value, dof, expected = chi2_contingency(df)

print("Chi-square:", chi2)
print("p-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

Conclusion: There is strong evidence to suggest that advertising wine during televised tennis matches is not effective.
