```python
import numpy as np
from scipy.stats import *
```

```python
WDUX = 0.15
WWTY = 0.19
WACO = 0.22
WTJW = 0.16
others = 0.28

n = 2000

expected = np.array([
    n * WDUX,
    n * WWTY,
    n * WACO,
    n * WTJW,
    n * others
])
observed = np.array([182, 536, 354, 151, 777])

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi2 statistic:", chi2_stat)
print("p-value:", p_value)
```

```python
df = len(expected) - 1 # k - 1
alpha = 0.10
critical_value = chi2.ppf(1 - alpha, df)
print("Critical value:", critical_value)
```

H0:
Current viewership shares are unchanged from the previous rating period.
This means the importance of channel views for each show remains the same.


Ha:
Current viewership shares differ from the viewership shares of the previous rating period.
This means that at least one category (channel or show) has a different viewership share.

```python
if p_value < alpha:
    print('strong evidence to reject H0 with alpha = {}'.format(alpha))
else:
    print('weak evidence to reject H0 with alpha = {}'.format(alpha))
```
