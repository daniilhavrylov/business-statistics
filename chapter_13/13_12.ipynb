```python
import numpy as np
from scipy.stats import *
```

```python
n = 60

data = np.array([
    34, 33, 33, 29, 26, 33, 28, 25, 32, 33,
    32, 25, 27, 33, 22, 27, 32, 33, 32, 29,
    24, 30, 20, 34, 31, 32, 30, 35, 33, 31,
    32, 28, 30, 31, 31, 33, 29, 27, 34, 31,
    31, 28, 33, 31, 32, 28, 26, 29, 32, 34,
    32, 30, 34, 32, 30, 30, 32, 31, 29, 33
])

alpha = 0.05

mean = np.mean(data)
std = np.std(data, ddof=1)

print('Sample mean:', mean)
print('Standard deviation', round(std, 4))
```

```python
interval1 = data < mean - 2 * std
interval2 = (data >= mean - 2 * std) & (data < mean - std)
interval3 = (data >= mean - std) & (data < mean)
interval4 = (data >= mean) & (data < mean + std)
interval5 = (data >= mean + std) & (data < mean + 2 * std)
interval6 = data >= mean + 2 * std

normal_intervals = [interval1, interval2, interval3, interval4, interval5, interval6]

for index, interval in enumerate(normal_intervals, start=1):
    print(f'Interval {index}: ', np.sum(interval))
```

We cannot use a chi-square test for this data because, for the test to be valid, each expected interval should contain at least 5 observations.
