```python
import numpy as np
from scipy.stats import *
```

```python
n = 1000

general_motors = 0.36
japanese = 0.26
ford = 0.21
chrysler = 0.09
other = 0.08

expected = np.array([
    n * general_motors,
    n * japanese,
    n * ford,
    n * chrysler,
    n * other
])
observed = np.array([193, 384, 170, 90, 163])

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi2 statistic:", chi2_stat)
print("p-value:", p_value)
```

```python
df = len(expected) - 1 # k - 1
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)
print("Critical value:", critical_value)
```

H0: Current new car market shares are no different from 1990 market shares.

Ha: Current new car market shares differ from those of 1990.
That is, at least one category has a different market share than it did in 1990.

```python
if chi2_stat > critical_value:
    print('strong evidence to reject H0 with alpha = {}'.format(alpha))
else:
    print('weak evidence to reject H0 with alpha = {}'.format(alpha))
```
