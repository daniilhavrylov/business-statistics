```python
import numpy as np
from scipy.stats import *
```

```python
observed = np.array([131, 273, 119, 301, 176, 200])
chi2_stat, p_value = chisquare(f_obs=observed)

print("Chi2 statistic:", chi2_stat)
print("p-value:", p_value)
```

Ho: The purchase frequencies are homogeneous across all six brands (no difference in buying probability).

Ha: The purchase frequencies are not homogeneous (at least one brand differs).

```python
alpha = 0.025

if alpha > p_value:
    print('strong evidence to reject H0 with alpha = {}'.format(alpha))
else:
    print('weak evidence to reject H0 with alpha = {}'.format(alpha))
```
